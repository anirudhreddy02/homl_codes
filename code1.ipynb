```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data_iris = load_iris() 
data_iris.keys()
```

```python
#print(data_iris.DESCR)
```

```python
data_iris.data[0:2], data_iris.target[0:2]
```

```python
df = pd.DataFrame(data_iris.data,columns=['a','b','c','d'])
df['e'] = 1
df.head(10)
```

```python
#X = data_iris.data
X = df.values
y= [0 if x == 0 else 1 for x in data_iris.target ]
```

```python
theta = np.random.rand(5,1)
theta
```

```python
X[0]
```

```python
ws = X@ theta
ws
```

```python
logit = 1/(1+np.exp(-ws))
logit
```

```python
y_hat_function = lambda x: 1 if x >=.5 else 0
y_hat = [y_hat_function(x) for x in logit ]
y_hat
```

```python
cost_fn = np.zeros((150,1))
for x in range(len(y_hat)):
    cost_fn[x] = (y_hat[x]* np.log([logit[x]]))+((1-y_hat[x])*np.log([1-logit[x]]))
cost_fn
```

```python
sum(-cost_fn)/150
```

```python
#grad_1 = (logit[0]-y[0])*X[0][0]
grad_1 = sum([(logit[i]-y[i])*X[i][0] for i in range(len(y))])
grad_2 = sum([(logit[i]-y[i])*X[i][1] for i in range(len(y))])
grad_3 = sum([(logit[i]-y[i])*X[i][2] for i in range(len(y))])
grad_4 = sum([(logit[i]-y[i])*X[i][3] for i in range(len(y))])
grad_5 = sum([(logit[i]-y[i])*X[i][4] for i in range(len(y))])
grad_1,grad_2,grad_3,grad_4,grad_5
```

```python
lr = 0.01 
theta[0] -= lr*grad_1
theta[1] -= lr*grad_2
theta[2] -= lr*grad_3
theta[3] -= lr*grad_4
theta[4] -= lr*grad_5
theta
```
